```python
from __future__ import print_function
import nltk
from nltk.collocations import *
from nltk import word_tokenize, sent_tokenize
import random

count = 0
countNode = 0
INT_MAX = 2147483647


randomList = []


def optimalSearchTree(keys, freq, n): 
    cost = [[0 for x in range(n)] for y in range(n)] 
  
    for i in range(n): 
        cost[i][i] = freq[i]  
        
    for L in range(2, n + 1): 
        for i in range(n - L + 2): 
            j = i + L - 1
            if i >= n or j >= n: 
                break
            cost[i][j] = INT_MAX 
            for r in range(i, j + 1): 
                c = 0
                if (r > i): 
                    c += cost[i][r - 1] 
                if (r < j): 
                    c += cost[r + 1][j] 
                c += sum(freq, i, j)  
                if (c < cost[i][j]): 
                    cost[i][j] = c 
    return cost[0][n - 1]  

def sum(freq, i, j):
    s = 0
    for k in range(i, j + 1): 
        s += freq[k]  
    return s


def bigram():
    NONE = " "
    f = open('WordCloud.txt')
    raw = f.read()
    #print(raw)
    #finder.ngram_fd.viewitems()
    tokens = nltk.tokenize.word_tokenize(raw)
    #tokens = nltk.word_tokenize(raw)
    bgs = nltk.bigrams(tokens)

    #Compute frequency distribution for all the bigrams in the text
    fdist = nltk.FreqDist(bgs)
    for k,v in fdist.items():
        print (k,v)
    return NONE







class Node(object):
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    # Inserts the number to the binary search tree
    def insert(self, data):
        global count
        count = count + 1
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
            
    # Counts the total number of nodes in the binary search tree  
    def nodeCounts(self, root):# count = 0):
        count = 1 
        if root is not None:
            if root.left is not None:
                count += self.nodeCounts(root.left) 
            if root.right is not None:
                count += self.nodeCounts(root.right)
        return count
    
    # Deletes a node in a binary search tree, if node to delete does not exist, do  nothing 
    def deleteNode(self, root, data): 
        # Base Case 
        if root is None: 
            return root  
        if data < root.data: 
            root.left = self.deleteNode(root.left, data) 
        elif(data > root.data): 
            root.right = self.deleteNode(root.right, data) 
        else: 
            # Node with only one child or no child 
            if root.left is None : 
                temp = root.right  
                root = None 
                return temp       
            elif root.right is None : 
                temp = root.left  
                root = None
                return temp 
            temp = minValueNode(root.right) 
            # Copy the inorder successor's content to temp
            root.data = temp.data
            # Delete the inorder successor on right side
            root.right = self.deleteNode(root.right , temp.data)
        return root
    
    # This prints the contents in the tree
    def print_tree(self):
        if self.left:
            self.left.print_tree()
        print(self.data, end=" ")
        if self.right:
            self.right.print_tree()
        return " "
    
    # This gets the height of the tree
    def height(self, root):
        if root is None:
            return 0
        else:
            return max(self.height(root.left), self.height(root.right)) + 1
    

print("Opening WordCloud.txt to store in binary search tree")
print("Everything is functional. It may not work correctly but it compiles and runs.")
fileChoose = "WordCloud.txt"
#fileChoose = input("Enter file name: ")
textFile = open(fileChoose, "r")
arrays = open(fileChoose).read().split()
arrays = [str(numbers) for numbers in arrays]


root = Node(arrays[0]) 
for i in range(1,len(arrays)):
    root.insert(arrays[i])
inputChoice = 0
while inputChoice != -1:
    
    print("\nHere is the choice menu. Pick a choice below or enter -1 to exit")
    print("1:   Insert to the optimal binary search tree")
    print("2:   Delete a node from the optimal binary search tree")
    print("3:   Return the height of the optimal binary search tree")
    print("4:   Return the number of nodes from the optimal binary search tree")
    print("5:   Print the values that is in the optimal Binary Search Tree")
    print("6:   Return the cost of the optimal binary search tree.")
    print("7.   Shuffle the words in the text file to randomize the order.")
    print("8.   Get the bigrams of the words from the textfile.")
    inputChoice = int(input("Enter your choice number: "))
    if inputChoice == 1:
        nodeInput = 0
        print("\nWhat number/word would you like to insert into the binary search tree?")
        nodeInput = str(input("Number/Word: "))
        root.insert(nodeInput)
    if inputChoice == 2:
        nodeDelete = 0
        print("\nWhat number/word would you like to delete from the binary search tree?")
        nodeDelete = str(input("Number/Word: "))
        root.deleteNode(root, nodeDelete)
    if inputChoice == 3:
        print("\nThe height of the binary search tree is:", root.height(root))
    if inputChoice == 4:
        print("\nThe number of nodes in the binary search tree is:", root.nodeCounts(root))
    if inputChoice == 5:
        print("\n", root.print_tree())  
    if inputChoice == 6:
        arraysLength = len(arrays)
        for i in range (0, arraysLength):
            x = random.randint(1,50)
            randomList.append(x)
        print("Cost of the Optimal Binary Search Tree is: ", optimalSearchTree(arrays, randomList, arraysLength))
    if inputChoice == 7: 
        shuffle = []
        for i in range(0,len(arrays)):
            shuffle.append(arrays[i])
        # The shuffle works and is functional. 
        print("\nThis is arrays: ", arrays)
        random.shuffle(shuffle) 
        print("\nThis is the new arrays shuffled: ", shuffle)
    if inputChoice == 8:
        print(bigram())
    if inputChoice == -1:
        print("\nGoodBye!") 
    if inputChoice == 0 or inputChoice > 8 or inputChoice < -1:
        print("\nThat is not a valid option")  
```
